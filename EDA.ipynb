```python
#getting all the categorical features in the dataset
categorical_features=[features for features in df1.columns if ((df1[features].dtypes=="O") & (features not in ['subscribed']))]
categorical_features
```

```python
#finding the number of unique values of each categorical feature
for feature in categorical_features:
    print('The feature is {}, and the number of categories are {}'.format(feature,len(df1[feature].unique())))
```

```python
#finding the counts of different categories present in each categorical variable
#representing the above graphs with values
for categorical_feature in categorical_features:
    print(categorical_feature,":")
    print()
    value_counts = df1[categorical_feature].value_counts()
    print(value_counts)
    print("\n\n\n")
```

```python
#finding the counts of different categories present in each categorical variable
#presenting graphically
plt.figure(figsize=(15,90), facecolor='white')
pltno=1
for categorical_feature in categorical_features:
    ax=plt.subplot(12,2,pltno)
    sns.countplot(y=categorical_feature,data=df1)
    plt.xlabel("categorical_feature")
    plt.title(categorical_feature)
    pltno+=1
plt.show()
```

```python
#finding the counts of different categories present in each categorical variable grouped by subscription
#representing the above graphs with values
for categorical_feature in categorical_features:
    print(categorical_feature,":")
    value_counts = df1.groupby(["subscribed"])[categorical_feature].value_counts()
    print(value_counts,"\n")
    print("\n")
```

```python
#finding the counts of different categories present in each categorical variable grouped by subscription
#presenting using graphs
for categorical_feature in categorical_features:
    sns.catplot(x='subscribed', col=categorical_feature, kind='count', data=df1)
plt.show()
```

```python
#numerical data in the dataset
numerical_features=[feature for feature in df1.columns if ((df1[feature].dtypes!="O") & (feature not in ['subscribed']))]
print('Number of numerical varibles :', len(numerical_features))
df1[numerical_features].head(10)
```

```python
#finding the discrete variables
discrete_feature=[feature for feature in numerical_features if len(df1[feature].unique())<25]
print("Discrete variables count:{}".format(len(discrete_feature)))
print(discrete_feature)
```

```python
#finding continuous features
continuous_features=[feature for feature in numerical_features if feature not in discrete_feature+['subscribed']]
print("Continuous feature count {}".format(len(continuous_features)))
print(continuous_features)
```

```python
#plotting the distribution plot for each continuous features
plt.figure(figsize=(20,60))
plotno=1
for continuous_feature in continuous_features:
    ax=plt.subplot(12,2, plotno)
    sns.distplot(df1[continuous_feature], kde=False,bins=5)
    plt.xlabel(continuous_feature)
    plotno+=1
plt.show()
```

```python
#plotting box plots for each continuous features and classified by the column subscribed
plt.figure(figsize=(20,90))
plotno=1
for feature in continuous_features:
    ax=plt.subplot(12,2, plotno)
    sns.boxplot(x="subscribed",y=df1[feature],data=df1)
    plt.title(feature)
    plotno+=1
plt.show()
```
